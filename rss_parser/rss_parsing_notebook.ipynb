```python
from bs4 import BeautifulSoup
def udn_rss_urls_extractor():
    udn_rss_url_dict = {}
    udn_rss_landing_url = 'https://udn.com/rssfeed/lists/1001'
    r = requests.get(udn_rss_landing_url, headers = headers)
    web_content = r.content
    soup = BeautifulSoup(web_content, 'lxml')
    for dt in soup.find_all('dt'):
        url_raw = dt.find('a')
        if url_raw and url_raw.get('href').find('rssfeed') != -1 and url_raw.text.strip():
            udn_rss_url_dict[url_raw.text.strip()] = url_raw.get('href')
    return udn_rss_url_dict
#udn_rss_urls_extractor()

def udn_rss_parser():
    udn_rss_url_dict = udn_rss_urls_extractor()
    for category, url in udn_rss_url_dict.items():

        r = requests.get(url, headers = headers)
        r.encoding = 'utf-8'
        web_content = r.text
        rss_source = '經濟日報'
        rss_category = category.strip()
        print(rss_source)
        print(rss_category)
        for index, match in enumerate(re.finditer(r'(\<link\>)(.+?)(<\/link\>)', web_content)):
            if index < 1:
                continue
            print(match.group(2))
udn_rss_parser()
```
